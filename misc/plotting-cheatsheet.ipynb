## 0. Setup — Understanding the Imports

Before ANY plotting, we need to load the toolkit. Here's the line you'll put at the top of every plotting script:

```python
import matplotlib.pyplot as plt
```

### Breaking it down:

```
import  matplotlib.pyplot  as  plt
  ↑           ↑            ↑    ↑
  1           2            3    4
```

| Part | What it means |
|------|---------------|
| 1. `import` | Python keyword: "load something from outside my script" |
| 2. `matplotlib.pyplot` | The thing being loaded (the plotting library's main module) |
| 3. `as` | "...but give it a different name" |
| 4. `plt` | The short nickname we'll use everywhere |

### Why we write it this way

**What is matplotlib?** It's Python's most popular plotting library — a "drawing toolkit" that turns numbers into charts. Install once: `pip install matplotlib`.

**What is `.pyplot`?** matplotlib is huge. The `.pyplot` part is the simple, beginner-friendly interface (the dot `.` means "the pyplot section INSIDE matplotlib").

**Why `as plt`?** Without it, you'd have to type `matplotlib.pyplot.plot(...)` every time. Way too long. So everyone — tutorials, books, Stack Overflow — uses `plt` as a convention.

### Common aliases you'll see everywhere

```python
import numpy as np              # NumPy → np
import pandas as pd             # Pandas → pd
import matplotlib.pyplot as plt # matplotlib.pyplot → plt
import torch                    # PyTorch (no alias — already short)
```

These are universal. Use them so your code looks like everyone else's.

### After `import matplotlib.pyplot as plt`, you can do:

```python
plt.plot(x, y)       # draw a line
plt.scatter(x, y)    # draw dots
plt.bar(cat, val)    # draw bars
plt.hist(data)       # draw histogram
plt.imshow(matrix)   # show a matrix
plt.title('...')     # add title
plt.xlabel('...')    # label x-axis
plt.ylabel('...')    # label y-axis
plt.legend()         # show legend
plt.grid(True)       # add grid
plt.show()           # display it
plt.savefig('f.png') # save to file
```

Every example in this notebook uses these. Now let's start plotting!

````python
# Load the libraries we need
import numpy as np                    # for number/array operations (aliased to 'np')
import matplotlib.pyplot as plt       # for plotting (aliased to 'plt')

# Reset matplotlib to default style (undoes any previous customizations)
plt.rcdefaults()

print("Ready to plot! numpy loaded as 'np', matplotlib.pyplot loaded as 'plt'")
````

## 1. Line Plot — The Bread and Butter

Used for: **loss curves, accuracy over time, learning rate schedules**.

````python
# Simple line plot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 4))        # width × height in inches
plt.plot(x, y1, label='sin(x)')    # line 1
plt.plot(x, y2, label='cos(x)')    # line 2
plt.xlabel('x')
plt.ylabel('y')
plt.title('Two simple lines')
plt.legend()                        # show labels
plt.grid(True, alpha=0.3)           # light grid
plt.show()
````

````python
# Line styles, colors, widths — control every detail
x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 4))

# format string: 'color + marker + linestyle'
plt.plot(x, np.sin(x), 'b-', linewidth=2, label='blue solid')
plt.plot(x, np.sin(x) + 0.5, 'r--', linewidth=2, label='red dashed')
plt.plot(x, np.sin(x) + 1.0, 'g:', linewidth=2, label='green dotted')
plt.plot(x, np.sin(x) + 1.5, 'k-.', linewidth=2, label='black dash-dot')

# Using hex colors and named parameters
plt.plot(x, np.sin(x) + 2.0, color='#FF6B35', linewidth=3, 
         linestyle='-', marker='o', markersize=3, markevery=10,
         label='orange with dots every 10 points')

plt.legend(loc='lower left', fontsize=9)
plt.title('Line styles reference')
plt.grid(True, alpha=0.3)
plt.show()

print("Color codes: 'b'=blue, 'r'=red, 'g'=green, 'k'=black, 'c'=cyan, 'm'=magenta, 'y'=yellow")
print("Line styles: '-'=solid, '--'=dashed, ':'=dotted, '-.'=dash-dot")
print("Markers:     'o'=circle, 's'=square, '^'=triangle, 'x'=cross, '*'=star")
````

## 2. Scatter Plot — Data Points

Used for: **visualizing datasets, predictions vs actuals, clustering**.

````python
# Scatter plot with colored classes (classification viz)
np.random.seed(42)

# Fake 2-class data
class_0 = np.random.randn(100, 2) + np.array([1, 1])
class_1 = np.random.randn(100, 2) + np.array([-1, -1])

plt.figure(figsize=(8, 6))
plt.scatter(class_0[:, 0], class_0[:, 1], 
            c='blue', alpha=0.6, s=50, marker='o', label='Class 0')
plt.scatter(class_1[:, 0], class_1[:, 1], 
            c='red', alpha=0.6, s=50, marker='s', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('2-class classification data')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')           # equal aspect ratio
plt.show()

print("Key params:")
print("  c = color (string or array of values for colormap)")
print("  s = size of points")
print("  alpha = transparency (0-1)")
print("  marker = shape (o, s, ^, x, *)")
````

````python
# Scatter plot with color mapped to a value (continuous)
# Use this when color ENCODES something (like prediction confidence)

x = np.random.randn(200)
y = np.random.randn(200)
values = x * y                    # some "measure" per point

plt.figure(figsize=(9, 6))
scatter = plt.scatter(x, y, c=values, cmap='coolwarm', s=80, 
                       edgecolor='black', linewidth=0.5)
plt.colorbar(scatter, label='x * y value')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter with color-mapped values (cmap)')
plt.grid(True, alpha=0.3)
plt.show()

print("Useful cmaps:")
print("  'viridis', 'plasma'   — perceptually uniform (good default)")
print("  'coolwarm', 'RdBu_r'  — diverging (for signed values)")
print("  'Blues', 'Reds'       — single-color (for positive values)")
print("  'tab10', 'Set2'       — discrete categories")
````

## 3. Multiple Subplots — Comparing Things Side by Side

Used for: **comparing experiments, showing train vs val, before vs after**.

````python
# 1 row, 2 columns — most common layout
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

x = np.linspace(0, 10, 100)

ax1.plot(x, np.sin(x), 'b-')
ax1.set_title('Sine')
ax1.set_xlabel('x')
ax1.grid(True, alpha=0.3)

ax2.plot(x, np.cos(x), 'r-')
ax2.set_title('Cosine')
ax2.set_xlabel('x')
ax2.grid(True, alpha=0.3)

plt.tight_layout()  # prevents overlap
plt.show()
````

````python
# Grid layout: 2 rows × 3 columns
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
# axes is a 2D numpy array — axes[row][col]

x = np.linspace(0, 10, 100)
functions = [
    (np.sin, 'sin', 'blue'),
    (np.cos, 'cos', 'red'),
    (np.tan, 'tan', 'green'),
    (np.exp, 'exp', 'orange'),
    (np.log, 'log', 'purple'),
    (lambda x: x**2, 'x²', 'brown'),
]

for ax, (func, name, color) in zip(axes.flatten(), functions):
    try:
        y = func(x)
        ax.plot(x, y, color=color, linewidth=2)
    except:
        pass
    ax.set_title(name)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Tips:")
print("  axes.flatten() → loops through all subplots easily")
print("  axes[0, 1]     → access by [row, col] for 2D grid")
print("  tight_layout() → always add at the end to fix spacing")
````

## 4. Log Scale — When Values Span Many Orders of Magnitude

Used for: **loss curves (loss drops from 10 → 0.001), exponential growth, timing comparisons**.

````python
# Same data, linear vs log scale
epochs = np.arange(1, 100)
loss = 10 * np.exp(-epochs / 20)   # exponential decay (typical loss)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

ax1.plot(epochs, loss, 'b-', linewidth=2)
ax1.set_title('Linear scale — late improvements look tiny')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.grid(True, alpha=0.3)

ax2.plot(epochs, loss, 'b-', linewidth=2)
ax2.set_yscale('log')  # ← THE MAGIC LINE
ax2.set_title('Log scale — see improvements at every level')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss (log scale)')
ax2.grid(True, alpha=0.3, which='both')  # major + minor grid

plt.tight_layout()
plt.show()

print("Rule of thumb: if loss changes by more than 10x,")
print("always include a log-scale version of the plot.")
````

## 5. Contour Plot — 2D Functions, Decision Boundaries

Used for: **loss landscapes, classifier decision boundaries, 2D function visualization**.

````python
# Contour plot — showing a 2D function

# Step 1: make a grid
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)     # both shapes: (100, 100)

# Step 2: compute the function value at every grid point
Z = X**2 + Y**2               # a bowl shape

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Line contours
c1 = ax1.contour(X, Y, Z, levels=10, cmap='viridis')
ax1.clabel(c1, inline=True, fontsize=8)  # labels on contour lines
ax1.set_title('contour() — line contours')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True, alpha=0.3)

# Filled contours  
c2 = ax2.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(c2, ax=ax2)
ax2.set_title('contourf() — filled contours')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.tight_layout()
plt.show()

print("contour()  → just lines (use clabel() to label them)")
print("contourf() → filled regions with colormap")
print("Use both together for maximum clarity.")
````

## 6. Bar Chart — Counts & Frequencies

Used for: **token frequencies, class distributions, comparing metrics**.

````python
# Simple bar chart
labels = ['Model A', 'Model B', 'Model C', 'Model D']
accuracies = [72, 81, 89, 94]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

# Vertical bars
bars = ax1.bar(labels, accuracies, color='steelblue', edgecolor='black')
ax1.set_ylabel('Accuracy (%)')
ax1.set_title('Vertical bar chart')
ax1.set_ylim(0, 100)
ax1.grid(True, alpha=0.3, axis='y')

# Add value labels on top of bars
for bar, val in zip(bars, accuracies):
    ax1.text(bar.get_x() + bar.get_width()/2, val + 1, 
             f'{val}%', ha='center', fontweight='bold')

# Horizontal bars — better for long labels
ax2.barh(labels, accuracies, color='tomato', edgecolor='black')
ax2.set_xlabel('Accuracy (%)')
ax2.set_title('Horizontal bar chart')
ax2.grid(True, alpha=0.3, axis='x')

plt.tight_layout()
plt.show()
````

````python
# Grouped bar chart — comparing multiple series
categories = ['Train', 'Val', 'Test']
before_finetune = [85, 72, 70]
after_finetune  = [92, 89, 88]

x = np.arange(len(categories))
width = 0.35

fig, ax = plt.subplots(figsize=(9, 5))
ax.bar(x - width/2, before_finetune, width, label='Before fine-tuning', color='steelblue')
ax.bar(x + width/2, after_finetune,  width, label='After fine-tuning',  color='tomato')

ax.set_ylabel('Accuracy (%)')
ax.set_title('Model performance: before vs after fine-tuning')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()
ax.grid(True, alpha=0.3, axis='y')
ax.set_ylim(0, 100)

plt.tight_layout()
plt.show()
````

## 7. Heatmap — Matrices & Attention Weights

Used for: **attention visualization, confusion matrices, correlation matrices, weight inspection**.

This is HUGE for LLMs — attention is a matrix you'll visualize constantly.

````python
# Heatmap — visualize any 2D matrix
# Example: fake "attention" matrix (which word attends to which)

words = ['The', 'cat', 'sat', 'on', 'the', 'mat']
n = len(words)
np.random.seed(0)

# Fake attention — higher values = more attention
attention = np.random.rand(n, n)
# Make diagonal strong (words attend to themselves)
attention = attention + np.eye(n) * 0.5
# Normalize rows to sum to 1 (like real attention)
attention = attention / attention.sum(axis=1, keepdims=True)

fig, ax = plt.subplots(figsize=(8, 7))

im = ax.imshow(attention, cmap='Blues', aspect='auto')
plt.colorbar(im, ax=ax, label='Attention weight')

# Add ticks with word labels
ax.set_xticks(range(n))
ax.set_yticks(range(n))
ax.set_xticklabels(words)
ax.set_yticklabels(words)

# Add text annotations in each cell
for i in range(n):
    for j in range(n):
        text_color = 'white' if attention[i, j] > 0.25 else 'black'
        ax.text(j, i, f'{attention[i, j]:.2f}', 
                ha='center', va='center', color=text_color, fontsize=9)

ax.set_xlabel('Key (looking at)')
ax.set_ylabel('Query (from)')
ax.set_title('Attention weights heatmap')

plt.tight_layout()
plt.show()

print("This is exactly how you'll visualize attention in transformers!")
print("Key: what each word is 'attending to' when processing the other.")
````

## 8. 3D Surface Plot — Loss Landscapes

Used for: **visualizing loss over 2 parameters, understanding optimization**.

````python
# 3D surface plot — great for loss landscape visualization
from mpl_toolkits.mplot3d import Axes3D

# Create grid
x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y) + 0.5 * X**2 * 0.2  # bumpy landscape

fig = plt.figure(figsize=(12, 5))

# Surface plot
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8, edgecolor='none')
ax1.set_title('3D surface — the loss "landscape"')
ax1.set_xlabel('w1')
ax1.set_ylabel('w2')
ax1.set_zlabel('Loss')

# Corresponding 2D contour (often more readable)
ax2 = fig.add_subplot(122)
c = ax2.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(c, ax=ax2)
ax2.set_title('Same data as 2D contour — easier to read')
ax2.set_xlabel('w1')
ax2.set_ylabel('w2')

plt.tight_layout()
plt.show()

print("3D plots look cool but are often HARDER to read than contours.")
print("Use 3D for wow factor; use contours for clarity.")
````

## 9. Histogram — Distributions

Used for: **weight distribution, activation distributions, data exploration**.

````python
# Histogram showing distribution of values

np.random.seed(0)
weights_before = np.random.randn(1000) * 0.5       # initialization
weights_after  = np.random.randn(1000) * 0.5 + 0.2 # after training (biased)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

ax1.hist(weights_before, bins=50, color='steelblue', edgecolor='black', alpha=0.7)
ax1.set_title('Weights at initialization')
ax1.set_xlabel('Weight value')
ax1.set_ylabel('Count')
ax1.axvline(x=0, color='red', linestyle='--', label='mean = 0')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Overlapping histograms for comparison
ax2.hist(weights_before, bins=50, color='steelblue', alpha=0.5, label='Before training')
ax2.hist(weights_after,  bins=50, color='tomato',    alpha=0.5, label='After training')
ax2.set_title('Weight distribution: before vs after')
ax2.set_xlabel('Weight value')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Useful for spotting issues:")
print("  Weights all near 0        → not learning")
print("  Weights exploding to huge → need gradient clipping")
print("  Weights all positive      → possible activation issue")
````

## 10. Styling — Make Your Plots Publication-Ready

````python
# Annotations, highlighted regions, styling

x = np.linspace(0, 100, 100)
train_loss = 3 * np.exp(-x / 20) + np.random.randn(100) * 0.05
val_loss   = 3 * np.exp(-x / 25) + 0.3 + np.random.randn(100) * 0.05

# Make validation go up after epoch 60 (overfitting)
val_loss[60:] += np.linspace(0, 0.8, 40)

best_epoch = np.argmin(val_loss)

fig, ax = plt.subplots(figsize=(11, 5))

ax.plot(x, train_loss, 'b-', linewidth=2, label='Train loss')
ax.plot(x, val_loss,   'r-', linewidth=2, label='Val loss')

# Mark the best epoch
ax.axvline(x=best_epoch, color='green', linestyle='--', linewidth=2, 
           label=f'Best epoch ({best_epoch})')

# Shade the overfitting region
ax.axvspan(best_epoch, 100, alpha=0.2, color='red', label='Overfitting zone')

# Annotation with an arrow
ax.annotate('Overfitting starts here!',
            xy=(best_epoch, val_loss[best_epoch]), 
            xytext=(best_epoch + 15, val_loss[best_epoch] + 0.5),
            fontsize=11, fontweight='bold', color='red',
            arrowprops=dict(arrowstyle='->', color='red', lw=2))

ax.set_xlabel('Epoch', fontsize=12)
ax.set_ylabel('Loss', fontsize=12)
ax.set_title('Training progress with annotations', fontsize=14, fontweight='bold')
ax.legend(loc='upper right', fontsize=10)
ax.grid(True, alpha=0.3)

# Trim unused whitespace
ax.set_xlim(0, 100)

plt.tight_layout()
plt.show()

print("Useful functions:")
print("  ax.axvline / axhline → vertical / horizontal lines")
print("  ax.axvspan / axhspan → shaded regions")
print("  ax.annotate          → text + arrows")
print("  ax.set_xlim / ylim   → control the axis range")
````

### When to use what
| Plot | Use case |
|------|---------|
| **Line** | Loss curves, metrics over time |
| **Scatter** | Data points, predictions vs targets |
| **Bar** | Comparing categories/models |
| **Histogram** | Distribution of values |
| **Contour** | 2D functions, decision boundaries |
| **Heatmap (imshow)** | Attention weights, confusion matrix |
| **3D surface** | Loss landscapes (but contour usually clearer) |
